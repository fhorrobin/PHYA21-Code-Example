```python
# Some imports we need to plot and calculate stuff.
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Here we plot the two functions, in blue is the small angle approximation result and in orange is the actual function. We can see that the two functions are very close to one another for $\theta < 1$

```python
theta_max = 2 * np.pi
theta = np.linspace(0, theta_max, 100)

y1 = theta**3
y2 = np.sin(theta)**2 * np.tan(theta)

plt.figure(dpi=150)
plt.plot(theta, y1)
plt.plot(theta, y2, '--')
plt.xlabel(r"$\theta$ (radians)")
plt.ylabel(r"$y(\theta)$")
```

```python
theta_max = 0.04
theta = np.linspace(0, theta_max, 100)

#The actual value of the constant on the RHS.
val = (8.99e9 * 5e-9**2) / (3e-3 * 9.8)

y1 = theta**3 - val
y2 = np.sin(theta)**2 * np.tan(theta) - val
```

Then we can evaluate the value of the constant and plot the expressions: 
$$ y_1 (\theta) - c = 0 $$
$$ y_2 (\theta) - c = 0 $$

Where $c$ is the value of the constant calculated above

$$c \approx \frac{k q^2}{mg} $$.

The solution we want is the root of these expressions, or the value when $y = 0$. We see that graphicall this is pretty much exactly the same for both of the solutions, showing that for this problem the small angle approximation works incredibly well.

```python
plt.figure(dpi=150)
plt.plot(theta, y1)
plt.plot(theta, y2, '--')
plt.plot([0, theta_max], [0, 0])
plt.xlabel(r"$\theta$ (radians)")
plt.ylabel(r"$y(\theta)$")
```
